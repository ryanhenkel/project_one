```python
#Dependencies
import pandas as pd
from pathlib import Path
```

```python
#Store filepath in a variable
median_income = Path("Median_Income.csv")
vaccination_rate = Path("Vaccination_Rates.csv")
```

```python
#Read our Median Income data with the Pandas library
median_income_df = pd.read_csv(median_income)
median_income_df
```

```python
#Read our Vaccination Rate data with the Pandas Library
vaccination_rate_df = pd.read_csv(vaccination_rate)
vaccination_rate_df
```

```python
#Merge the datasets, 
la_data_complete = pd.merge(vaccination_rate_df, median_income_df, right_on = 'NEIGHBORHOOD', left_on = 'CITY')
la_data_complete
```

```python
#Sort the cities by 'Case Rate' in descending order
case_rate_df = la_data_complete.sort_values("Case_Rate", ascending=False)
case_rate_df
```

```python
#Sort the cities by 'Death Rate' in descending order
death_rate_df = la_data_complete.sort_values("Death_Rate", ascending=False)
death_rate_df.head(173)
```

```python
#Sort the cities by 'Percent Fully Vaccinated' in descending order
percent_fully_vaccinated_df = la_data_complete.sort_values("Percent_of_People_Fully_Vaccinated", ascending=False)
percent_fully_vaccinated_df
```
